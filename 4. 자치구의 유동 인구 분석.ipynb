```python
df = pd.read_csv('data/스마트서울 도시데이터 센서(S-DoT) 유동인구 측정 정보.csv', encoding='ansi', parse_dates=['측정시간'])
```

```python
df
```

```python
# 필요없는 컬럼 삭제
df.drop(['등록일'], axis =1, inplace=True)
```

```python
df.info()
```

### 시간대 분석을 위한 데이터 형식 바꾸기

```python
# 시간대 분석을 위해 측정시간(object)를 time(datetime) 형식으로 변형
df['time'] = pd.to_datetime(df['측정시간'], format='%Y-%m-%d_%H:%M:%S', errors='raise')
df
```

```python
# 새로운 측정시간 삽입을 위해 기존 측정시간컬럼 삭제
df.drop(['측정시간'], axis =1, inplace=True)
```

```python
# 측정시간 문자열분리
time = df['time'].dt.hour
```

```python
# 새로운 측정시간 컬럼 삽입
df['측정시간'] = time
df
```

```python
df.info()
```

```python
# 사용할 데이터 항목의 결측치 확인
df['time'].isnull().sum()
df['방문자수'].isnull().sum()
```

```python
# 자치구 이름 확인
df['자치구'].unique()
```

### 강남구의 시간대별 유동인구

```python
# 강남구에 속하는 방문자수의 합을 측정시간별로 구함
g20 = pd.pivot_table(data = df[df['자치구'] == 'Gangseo-gu'], 
                     index = '측정시간', aggfunc = 'sum').reset_index()
g20['방문자수'].plot.bar()
```

### 송파구의 시간대별 유동인구

```python
# 송파구에 속하는 방문자수의 합을 측정시간별로 구함
g20 = pd.pivot_table(data = df[df['자치구'] == 'Songpa-gu'], 
                     index = '측정시간', aggfunc = 'sum').reset_index()
g20['방문자수'].plot.bar()
```

### 전체 자치구의 시간대별 유동인구

```python
# 전체 자치구의 시간대별 유동인구 분석에 불필요한 컬럼들 삭제
df.drop(['time','모델번호','시리얼','행정동','지역'], axis =1, inplace=True)
df
```

```python
# 측정시간으로 그룹화 하여 방문자수 합을 구함
groups = df.groupby(['측정시간']).sum()
groups.sort_index()
```

```python
# 전체 자치구의 시간대별 유동인구합을 구함
total_time = groups.sort_index().plot.bar(ylabel = '유동인구합', color ='g', figsize=(5,5))
total_time
plt.title('전체 자치구 시간대별 유동인구')
```
